```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('/content/drive/MyDrive/Kaggle/080228-master/실습11_3/train_SectionDataset.csv' ,header=0)
df1 = df1[ df1['section'] == 0]
print( df1.shape )
print( df1['title'][0] )
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/Kaggle/080228-master/실습11_3/test_SectionDataset.csv' ,header=0)
pre_title = df2['title'][0]

#df1 = df1.append( pre_item, ignore_index=True )

pre_item = pd.DataFrame({
   'title': [pre_title],
   'section': [0]
})

df1 = pd.concat([df1, pre_item], ignore_index=True)

print( df1.shape )
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer( ngram_range=(1,3) )
c_vector_titles = count_vector.fit_transform( df1['title'] )
print( c_vector_titles.shape )
print( c_vector_titles[0] )
```

```python
from sklearn.metrics.pairwise import cosine_similarity

titles_c_sim = cosine_similarity( c_vector_titles )
print( titles_c_sim )

titles_c_sim = titles_c_sim.argsort()[:, ::-1]
print( titles_c_sim )
```

```python
sim_idx = titles_c_sim[2025, :6]
print(sim_idx)

print(df1.iloc[sim_idx])
```
